Importando módulos necessários.
imageio para leitura das imagens
matplotlib para exibição

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

Recebendo o nome do arquivo a ser tratado

```python
filename = str(input())
```

Abrindo a imagem

```python
img = imageio.imread(filename)
```

Define o método de exibição inline e o tamanho que as imagens serão exibidas.

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [18, 18]
```

Exibe imagem original

```python
plt.imshow(img, cmap="gray")
```

Recebe os parâmetros para o filtro de suavização (método, tamanho do filtro e sigma, desconsiderado caso o filtro não seja gaussiano).

Para a escolha do método de filtragem, digitge um dos números:
    1 - filtro de média
    2 - filtro de mediana
    3 - filtro gaussiano

```python
smooth_method = int(input())
smooth_fsize = int(input())
smooth_sigma = int(input())
```

```python
img_out = ft.smoothing_filter(img, method=smooth_method, n=smooth_fsize, sigma=smooth_sigma)
```

Prepara a exibição da imagem suavizada à esquerda.

Recebe o método de realce de bordas.
    1 - filtro laplaciano
    2 - filtro sobel

```python
sharpen_method = int(input())
```

```python
img_out = ft.sharpening_filter(img, method=sharpen_method)
```

Recebe o método de equalização da imagem.
    1 - Normalização
    2 - Equalização de histograma

```python
equalization_method = int(input())
```

Normaliza a imagem sempre. E caso a opção não seja equalização, chama a operação de equalização.

```python
if(equalization_method != 0):
    img_out = eq.equalize(img_out, 1)
if(equalization_method != 1):
    img_out = eq.equalize(img_out, equalization_method)
```

Exibe a imagem melhorada.

```python
plt.imshow(img_out, cmap="gray")
```

Recebe o método de colorização da imagem.
Os métodos de colorização 0, 1, 2 são específicos para bagagem.
Os métodos 3, 4, 5 podem ser usados para qualquer imagem.

```python
color_method = int(input())
```

Executa colorização

```python
img_out = cl.colorize_image(img_out.astype(np.uint8), color_method)
```

Exibe a imagem colorida.

```python
plt.imshow(img_out, cmap="gray")
```

Exibe as imagens. Original à esquerda, imagem tratada ao meio e imagem colorida à direita.
